```python
from matplotlib import pyplot as plt
import pickle
import pandas as pd
import numpy as np
```

```python
# loading notes csv
labels = pd.read_csv('labels.csv',delimiter="\t")
all_labels=list(labels["Label"])
```

```python
with open("no_correction_all_phases.pkl",'rb') as f:
    all_phases=pickle.load(f)
```

```python
plt.figure(figsize=(10,5))
plt.plot(np.arange(0,len(all_phases[915e6][0.836])//1),np.rad2deg(np.array([x for idx,x in enumerate(all_phases[915e6][0.836]) if idx%1==0])), 'o')
plt.xticks(np.arange(0,len(all_phases[915e6][0.836])//1), [x for idx,x in enumerate(all_labels) if idx%1==0],rotation=45, ha='right')
plt.ylim([0, 180])
plt.grid()
```
